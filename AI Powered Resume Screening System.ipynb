**Cleaning Data:**
1.URL,
2.Hashtag,
3.mentions,
4.special letters,
5.punchuation

```python
import re
def cleanResume(txt):
  cleanTxt = re.sub('http\S+\s','',txt)
  cleanTxt = re.sub('RT|cc',' ',cleanTxt)
  cleanTxt = re.sub('#\S+',' ',cleanTxt)
  cleanTxt = re.sub('@\S+',' ',cleanTxt)
  cleanTxt = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""),' ',cleanTxt)
  cleanTxt = re.sub(r'[^\x00-\x7f]',' ',cleanTxt)
  cleanTxt = re.sub('\s+',' ',cleanTxt)
  return cleanTxt
```

```python
cleanResume("My website link is this http://helloworld and access it")
```

```python
df['Resume'] = df['Resume'].apply(lambda x: cleanResume(x))
```

```python
df['Resume'][0]
```

**Word into categorical values**

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
le.fit(df['Category'])
df['Category']=le.transform(df['Category'])
```

**Vactorization**

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf=TfidfVectorizer(stop_words='english')
tfidf.fit(df['Resume'])
requiredTaxt=tfidf.transform(df['Resume'])
```

```python
requiredTaxt
```

**Splitting**

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(requiredTaxt,df['Category'],test_size=0.2,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

**Now let's train the model and print the classification report:**

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn.metrics import accuracy_score

clf=OneVsRestClassifier(KNeighborsClassifier())
clf.fit(x_train,y_train)
ypred=clf.predict(x_test)
print(accuracy_score(y_test,ypred))
```
